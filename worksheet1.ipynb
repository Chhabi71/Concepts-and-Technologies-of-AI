# Problem - 4: Matrix Operations:
• For the following arrays:

A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);

Prove following with Numpy:

```python
A =np.array(([3,4],[7,8]))
B =np.array(([5,3],[2,1]))
```

1. Prove A.A−1 = I.

```python
A_inv = np.linalg.inv(A)
A1= np.dot(A,A_inv)
A1= np.round(A1,decimals=0)
print("A * A^-1(rounded):\n", A1)
```

2. Prove AB ̸= BA.

```python
AB =np.dot(A,B)
BA =np.dot(B,A)
print("AB =\n",AB)
print("BA =\n",BA)
print("\n AB ̸= BA, Hence proved")
```

3. Prove (AB)T = BTAT



```python
LHS= (AB).T
RHS = np.dot(B.T, A.T)
print("(AB)^T:\n", LHS)
print("B^T A^T:\n", RHS)
print("\n(AB)^T = B^T A^T, Hence proved")
```
